```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins_df = sns.load_dataset('penguins')

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
display(penguins_df.head())
```

```python
print("\nDataset Information:")
penguins_df.info()

print("\nDescriptive Statistics:")
display(penguins_df.describe())
```

```python
print("\nMissing values per column:")
display(penguins_df.isnull().sum())

plt.figure(figsize=(10, 6))
sns.heatmap(penguins_df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
print("\nPenguin count per island:")
display(penguins_df['island'].value_counts())
```

```python
plt.figure(figsize=(8, 6))
# Calculate value counts
island_counts = penguins_df['island'].value_counts()
sns.barplot(x=island_counts.index, y=island_counts.values, hue=island_counts.index, palette='viridis', legend=False)

# Add values on top of the bars
for index, value in enumerate(island_counts.values):
    plt.text(index, value + 0.5, str(value), ha='center', va='bottom')

plt.title('Penguin Count per Island')
plt.xlabel('Island')
plt.ylabel('Number of Penguins')
plt.show()
```

```python
print("\nGenerating pairplot...")
sns.pairplot(penguins_df, hue='species', palette='viridis')
plt.suptitle('Pairplot of Penguin Data by Species', y=1.02) # Adjust title position
plt.show()
```

```python
print("\nGenerating distribution plots for numerical features...")
numerical_cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=penguins_df, x=col, kde=True, hue='species', palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
print("\nGenerating count plots for categorical features...")
categorical_cols = ['species', 'sex', 'island']

plt.figure(figsize=(15, 5))
for i, col in enumerate(categorical_cols):
    if col != 'island': # 'island' already has a bar plot
        plt.subplot(1, 2, i + 1)
        sns.countplot(data=penguins_df, x=col, hue='species' if col != 'species' else None, palette='viridis')
        plt.title(f'Count of Penguins by {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
